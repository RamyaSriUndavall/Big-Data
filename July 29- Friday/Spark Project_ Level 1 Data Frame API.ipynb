## LEVEL-1  DATA FRAME API
1.    Shows the names and height by adding 1 to the height column.
2.    shows the player name and simultaneously checks whether or not they have height >170
3.    Show FIFA Popular Name and 0 or 1 depending on Height>170
4.    name of  shortest player 
5.    who is tallest of all. First we find the value of maximum height and then get the details of that player
6.    average height of the players in Argentina team.

```
from pyspark.sql import SparkSession
```

```
sc=SparkSession.builder.appName('Level1').getOrCreate()
```

```
df=sc.read.csv('/FileStore/tables/wc2018_players.csv',header=True,inferSchema=True)
```

```
df.printSchema
```

```
df.columns
```

```
df.show(5)
```

```
#1. Shows the names and height by adding 1 to the height column.
df_a=df.withColumn('Height',df.Height+1)
df_a.select('FIFA Popular Name','Height').show()
```

```
df.show(5)
```

```
#2. shows the player name and simultaneously checks whether or not they have height >170
from pyspark.sql.functions import when
df_b=df.select('FIFA Popular Name','Height')
#df_b=df_b.withColumn('Height',when(df['Height']>170,'yes').when(df['Height']>170,'no'))
df_b=df_b.withColumn('Height',when(df['Height']>170,'yes').otherwise('no'))
df_b.show()
```

```
#3. Show FIFA Popular Name and 0 or 1 depending on Height>170

df_c=df.withColumn('Height',when(df['Height']>170,1).otherwise(0))
```

```
df_c.select('FIFA Popular Name','Height').show(5)
```

```
#4. name of shortest player
df.orderBy('Height').select('FIFA Popular Name').show(1)
```

```
df.filter(df['Height']==min('Height')).show()
```

```
df.min('Height')
```

```
df.describe('Height').filter(df['summary']==min).show()
```

```
n=df.selectExpr('min(Height)')
n.show()
```

```
df.describe('Height').show()
df['Height'].min()
```

```
m=df.agg({'Height':'min'}).collect()
type(m)
```

```
m[0]
```

```
print(m[0][0]) # Is there a better way to find the min , and then store it ?
```

```
df.filter(df['Height']==m[0][0]).show()
```

```
df.orderBy('Height').show()
```

```
#6. average height of the players in Argentina team
df.groupBy('Team').mean('Height').filter(df['Team']=='Argentina').show()
```

```
2.# How many movies are released each year?

sc.sql("select count(movieId), year from Movies group by year").show(5)
```

```
#3.  How many number of movies are there for each rating?
sc.sql("select count(movieId),rating from Ratings group by rating").show()
```

```
#4. How many users have rated each movie?
sc.sql("select count( distinct userId), movieId from Ratings group by movieId order by MovieId").show()
```

```
#5. What is the total rating for each movie?

#sc.sql("select * from ratings").show()
#sc.sql("select count(rating), movieId from Ratings group by movieId order by movieId").show()
sc.sql("select sum(rating), movieId from Ratings group by movieId order by movieId").show(5)
```

```
#6. What is the average rating for each movie?
sc.sql("select avg(rating), movieId from Ratings group by movieId order by movieId").show()
```
