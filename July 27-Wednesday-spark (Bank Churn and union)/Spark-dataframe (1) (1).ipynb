```
stock=sc.read.csv("/FileStore/tables/appl_stock.csv",header=True,inferSchema=True)
```

```
stock.show(10)
```

```
stock.printSchema()
```

```
from pyspark.sql.functions import format_number
```

```
data=stock.select(stock['Date'],
             format_number(stock['High'].cast('float'),2).alias("HIGH"),
             format_number(stock['Open'].cast('float'),2).alias("OPEN"),
             format_number(stock['Low'].cast('float'),2).alias("LOW"),
             format_number(stock['Close'].cast('float'),2).alias("CLOSE"),
             format_number(stock['Volume'].cast('float'),2).alias("VOLUME"),
             format_number(stock['Adj Close'].cast('float'),2).alias("Adj Close")
                         )
```

```
data.show(10)
```

```
from pyspark.sql.functions import hours,dayofmonth,dayofweek,dayofyear,year,date_format,weekofyear,month
```

```
data.select(dayofmonth(data['Date'])).show()
```
